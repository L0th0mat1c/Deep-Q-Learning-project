# Grid Search Graphs

```python
import os
import pandas as pd
import numpy as np

import seaborn as sns
```

```python
PATH = os.getcwd()
DATA_PATH = os.path.join(PATH, "results_grid-search.csv")

df = pd.read_csv(DATA_PATH, header=0, on_bad_lines='skip', sep=';')

df = df.applymap(lambda x: float("{:.2f}".format(x)) if type(x) == float else x)
```

```python
df.head(3)
```

```python
df.groupby(["Epsilon"]).mean()
```

```python
sns.lineplot(data = df)
```

```python
sns.heatmap(df.groupby(["Epsilon"]).mean())
```

```python
sns.heatmap(df.groupby(["Alpha"]).mean())
```

```python
sns.heatmap(df.groupby(["Gamma"]).mean())
```
